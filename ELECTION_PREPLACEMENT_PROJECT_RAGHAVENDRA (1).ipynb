### Introduction:
The Analysis here is based on the election of Lok Sabha for the year 2019, Members of Lok Sabha or the lower house of India's Parliament are elected by being voted upon by all adult citizens of India, from a set of candidates who stand in their respective constituencies. Candidates who win the Lok Sabha elections are called 'Member of Parliament' and hold their seats for five years or until the body is dissolved by the President on the advice of the council of ministers.

### DATA CLEANING

```python
# Clean Assets and LIABILITIES,EDUCATION,CRIMINAL CASES Column values

def df_clean(x):
    try:
        x=x.split('\n')[0]
        x=x.replace('Rs ','')
        x=x.replace(',','')
        x=x.strip()
        return int(x)
    except:
        x=0
        return int(x)

df['ASSETS']=df['ASSETS'].apply((df_clean))
df['LIABILITIES'] = df['LIABILITIES'].apply((df_clean))
df['EDUCATION']=df['EDUCATION'].str.replace('Post Graduate\n','Post Graduate')
df['CRIMINAL CASES']=df['CRIMINAL CASES'].str.replace('Not Available','0')

#to change data type of CRIMINAL CASES from string to float 
df['CRIMINAL CASES']=df['CRIMINAL CASES'].astype(float)
df.head()
```

### Seperating categorical and numerical values

```python
# Seprating out the categorical and continuous variables-----> unique and nunique
def seperate_data_types(df):
    categorical = []
    numeric = []
    
    for column in df.columns:
        if(df[column].nunique())<40:
            categorical.append(column)
        else:
            numeric.append(column)
    return categorical,numeric
```

# MultiVariate Analysis

### What age requirements should be in place for gender-wise winning candidates?

```python
import warnings
warnings.filterwarnings('ignore')
fig, ax = plt.subplots(figsize= (5,4.8))
ax = sns.barplot(x=df['GENDER'],y=df['AGE'],hue=df['WINNER'],data=df,
                palette='magma',ci=False,edgecolor='black',linewidth=2)
for bar in ax.patches:                                             
    percentage =f"{round(bar.get_height() , 1)}"          
    x = bar.get_x() + bar.get_width() /2                           
    y = bar.get_height()                                            
    ax.annotate(percentage, (x, y), va = "bottom", ha ="center" ) 
plt.title('Gender wise Average Age of Winning Candidates')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

#### Interpretation:Average age of Male should be:55, Average age of Female should be:49
                        

###  What assets should the winning candidates of the Lok Sabha election have, gender wise?

```python

fig, ax = plt.subplots(figsize= (5,4.8))
ax = sns.barplot(x=df['GENDER'],y=df['ASSETS'],hue=df['WINNER'],data=df,
                palette='magma',ci=False,edgecolor='black',linewidth=2)
for bar in ax.patches:                                             
    percentage =f"{round(bar.get_height() , 1)}"          
    x = bar.get_x() + bar.get_width() /2                           
    y = bar.get_height()                                            
    ax.annotate(percentage, (x, y), va = "bottom", ha ="center" ) 
plt.title('Gender wise Average ASSETS of Winning Candidates')
plt.xlabel('Gender')
plt.ylabel('ASSETS')
plt.show()
```

#### Interpretation:Male should have 208038111 and Female should have 215822134 assest
